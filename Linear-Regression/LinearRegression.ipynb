```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('homeprices.csv')
df.head()
```

```python
plt.xlabel('Area(sq feet)')
plt.ylabel('Price(us $)')
plt.scatter(df.Area, df.Price, color='red', marker='+')
```

```python

reg = LinearRegression()
reg.fit(df[['Area']], df.Price)
```

```python

prec_price = reg.predict([[3300]])
print(f"Pridicted Price for input is {prec_price}$")
```

```python

reg.predict([[5000]])
```

```python

%matplotlib inline
plt.xlabel('Area',fontsize = 20)
plt.ylabel('Price', fontsize = 20)
plt.scatter(df.Area, df.Price, color='red', marker='+')
plt.plot(df.Area, reg.predict(df[['Area']]), color='blue')
```

```python
d = pd.read_csv('Area.csv')
d.head(3)
```

```python

price = reg.predict(d)
price
```

```python

d['prices'] = price
d
```

```python

d.to_csv('prediction.csv', index = False)
```

```python
# New input and its prediction
new_x = 3300
new_y = reg.predict([[new_x]])
# Plot data points
plt.scatter(df.Area, df.Price, color='blue')

# Plot regression line
plt.plot(df.Area, reg.predict(df[['Area']]), color='blue')

# Add crossing lines for the new point
plt.axvline(new_x, color='green')
plt.axhline(new_y, color='purple')

# Plot the new point
plt.scatter(new_x, new_y, color='orange')

# Labels and legend
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression with Crossing Lines')
```
